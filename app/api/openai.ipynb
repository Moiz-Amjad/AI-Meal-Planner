```python
from dotenv import load_dotenv
load_dotenv(dotenv_path=".env.local")
import os
from openai import OpenAI

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

```python
completion = client.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "I weigh 182 pounds and am 5'10. I want to lose weight. Please recommend me a wweekly meal plan to lose 5 pounds in a month. I work out 3 days a week with weightlifting training. Please return answer in JSON format"}
  ]
)
```

```python
message = completion.choices[0].message.content
print(message)
```
